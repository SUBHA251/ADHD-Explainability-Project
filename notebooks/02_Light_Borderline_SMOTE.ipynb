```python
from imblearn.over_sampling import BorderlineSMOTE
import pandas as pd

X_train = pd.read_csv("../data/X_train.csv")

y_train = pd.read_csv("../data/y_train.csv")
```

```python
print(X_train.shape)

print(y_train.shape)
```

```python
y_train = y_train.squeeze()

print(type(y_train))
```

```python
print("Before Light Borderline-SMOTE")

print(y_train.value_counts())
```

```python
smote = BorderlineSMOTE(
    sampling_strategy={1:300},
    random_state=42
)

X_train_resampled, y_train_resampled = smote.fit_resample(
    X_train,
    y_train
)
```

```python
print("After Light Borderline-SMOTE")

print(y_train_resampled.value_counts())
```

```python
print(y_train_resampled.value_counts())
```

```python
pd.DataFrame(
    X_train_resampled
).to_csv(
    "../data/X_train_resampled.csv",
    index=False
)

pd.DataFrame(
    y_train_resampled
).to_csv(
    "../data/y_train_resampled.csv",
    index=False
)

print("Balanced dataset saved")
```
