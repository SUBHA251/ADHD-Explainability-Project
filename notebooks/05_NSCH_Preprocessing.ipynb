```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/RAW/2023-2024_NSCH_Topical_CAHMI_DRC.csv")

print(df.shape)
```

```python
print(df["ADHD_2324"].value_counts(dropna=False))
```

```python
print(df["ADHD_2324"].value_counts(dropna=False))

# Keep only valid classes
df = df[df["ADHD_2324"].isin([1, 3])].copy()

# Binary target
df["ADHD_binary"] = df["ADHD_2324"].map({
    1: 0,  # No ADHD
    3: 1   # Current ADHD
})

print(df["ADHD_binary"].value_counts())
```

```python
adhd_cols = [
    "ADHD_2324",
    "ADHDSev_2324",
    "ADHDind_2324",
    "ADHDSevInd_2324",
    "ADHDMed_2324",
    "ADHDBehTreat_2324"
]

df.drop(columns=adhd_cols, errors="ignore", inplace=True)

print(df.shape)
```

```python
missing_codes = [90,95,96,97,98,99,9990,9999]

df = df.replace(missing_codes, np.nan)

print("Done")
```

```python
numeric_cols = df.select_dtypes(include=["number"]).columns

df[numeric_cols] = df[numeric_cols].replace(
    [90,95,96,97,98,99,9990,9999],
    np.nan
)

print("Done")
```

```python
missing_codes = [90,95,96,97,98,99,9990,9999]

for code in missing_codes:
    df = df.mask(df == code, np.nan)

print("Done")
```

```python
missing_pct = df.isnull().mean()

print(missing_pct.sort_values(ascending=False).head(20))
```

```python
keep_cols = missing_pct[missing_pct < 0.50].index

df = df[keep_cols]

print(df.shape)
```

```python
y = df["ADHD_binary"]

X = df.drop(
    columns=["ADHD_binary"],
    errors="ignore"
)

print(X.shape)
print(y.shape)
```
